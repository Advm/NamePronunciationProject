```python
dummydf.columns = ['word'] + allchars + ['OUT']
```

```python
dummydf
```

```python
shuffled = dummydf.sample(frac=1)
```

```python
allmodels = []
loss_arr =[(tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), "SCCE")]
size_arr=[32,64, 128] 
optimizer_arr =[(keras.optimizers.RMSprop(),"RMS"), (keras.optimizers.Adam(), "ADAM")] #, (keras.optimizers.Adadelta(), "ADADELTA"), (keras.optimizers.Adagrad(), "ADAGRAD")]
activation_arr = ['linear', 'relu', 'tanh']
layer_arr = [1,2,3]
for loss_object in loss_arr:
    for optimizers in optimizer_arr:
        for act in activation_arr:
            for size in size_arr:
                for layersx in layer_arr:
                    inputs = keras.Input(shape=(52,))
                    x = layers.Dense(size, activation=act)(inputs)
                    for i in range(layersx - 1):
                        x = layers.Dense(size, activation=act)(x)
                
                
                    outputs = layers.Dense(1)(x)

                    # Name of model based on its features



                    nameofmodel = str(size) +"x" + str(layersx) + "_Dense_" + loss_object[1] +"_" + optimizers[1] + "_" + act

                    # Connects the layers into a model
                    model = keras.Model(inputs=inputs, outputs=outputs, name= "%s" % nameofmodel) 

                    # Compiles the model so it is ready to be fitted
                    model.compile(loss=loss_object[0],
                                  optimizer=optimizers[0],
                                  metrics=['accuracy'])

                    # Adds the model to the array
                    allmodels.append(model)
```

```python
target = pd.DataFrame()
target['OUT'] = shuffled['OUT']
```

```python
target
```

```python
shuffled = shuffled.drop(['OUT'], axis = 1)
```

```python
allmodels[0]
```

```python
history = allmodels[0].fit(shuffled[:120000].to_numpy(), target[:120000].to_numpy())
```

```python
model = tf.keras.Sequential([
        # Shape [batch, time, features] => [batch, lstm_units]
        # Adding more `lstm_units` just overfits more quickly.
        tf.keras.layers.Dense(100, activation='relu'),
        
        tf.keras.layers.Dense(100,  activation='relu'),
        
        # Shape => [batch, out_steps*features]
        tf.keras.layers.Dense(1, activation = 'sigmoid'),
        ])
```

```python
model.compile(optimizer='Adam',
                 loss=tf.losses.BinaryCrossentropy(),
                 metrics=['accuracy'])
```

```python
X_train = shuffled.drop(['word'], axis=1)[:120000]
y_train = target[:120000]
X_test = shuffled.drop(['word'], axis=1)[120001:]
y_test = target[120001:]
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=1,validation_data=(X_test,y_test), verbose = 1)


```
